```python
import pandas as pd
import matplotlib.pyplot as plt

# Tworzenie danych
data = {
    'Produkt': ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D', 'Produkt E'],
    'Quantity': [150, 200, 50, 300, 100],
    'Price': [10.5, 20.0, 15.0, 5.0, 12.0]
}

# Tworzenie DataFrame
df = pd.DataFrame(data)

# Obliczanie całkowitej wartości sprzedaży dla każdego produktu
df['Total Sales'] = df['Quantity'] * df['Price']

# Wyświetlanie wyników
print(df[['Produkt', 'Total Sales']])

# Obliczanie średniej ceny sprzedawanych produktów
average_price = df['Price'].mean()

# Obliczanie całkowitej ilości sprzedanych produktów
total_quantity = df['Quantity'].sum()

# Wyświetlanie wyników
print(f"Średnia cena sprzedawanych produktów: {average_price:.2f}")
print(f"Całkowita ilość sprzedanych produktów: {total_quantity}")


# Wykres słupkowy przedstawiający całkowitą wartość sprzedaży dla każdego produktu
plt.figure(figsize=(10, 6))
plt.bar(df['Produkt'], df['Total Sales'], color='skyblue')
plt.title('Całkowita wartość sprzedaży dla każdego produktu')
plt.xlabel('Produkt')
plt.ylabel('Całkowita wartość sprzedaży')
plt.show()

# Wykres kołowy ilustrujący udział poszczególnych produktów w całkowitej wartości sprzedaży
plt.figure(figsize=(8, 8))
plt.pie(df['Total Sales'], labels=df['Produkt'], autopct='%1.1f%%', 
        colors=['skyblue', 'lightgreen', 'orange', 'pink', 'purple'])
plt.title('Udział poszczególnych produktów w całkowitej wartości sprzedaży')
plt.show()


```
