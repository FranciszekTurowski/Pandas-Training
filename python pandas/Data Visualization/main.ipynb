6. Narysuj wykres liniowy z trzema seriami danych: funkcji sinus, cosinus oraz ich sumy. Użyj wartości x od 0 do 10 (100 punktów). Oblicz wartości y za pomocą NumPy i narysuj je przy użyciu Matplotlib.

12. Narysuj wykres liniowy z błędami dla funkcji logarytmicznej. Dodaj losowe błędy do wartości y, używając wartości x od 0.1 do 10 (100 punktów). Wizualizację przeprowadź za pomocą Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generowanie danych
x = np.linspace(0.1, 10, 100)  # Wartości x od 0.1 do 10 (100 punktów)
y = np.log(x)  # Funkcja logarytmiczna
errors = np.random.normal(0, 0.2, size=len(x))  # Losowe błędy (średnia=0, odchylenie=0.2)

# Dodanie błędów do wartości y
y_with_errors = y + errors

# Tworzenie wykresu
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Funkcja logarytmiczna", color="blue")
plt.fill_between(x, y - errors, y + errors, color="blue", alpha=0.2, label="Błędy")
plt.scatter(x, y_with_errors, color="red", s=10, label="Dane z błędami")

# Dodanie tytułu i etykiet
plt.title("Wykres funkcji logarytmicznej z błędami")
plt.xlabel("x")
plt.ylabel("y = log(x)")
plt.legend()
plt.grid(True)
plt.show()
```

13. Stwórz wykres radarowy dla czterech zmiennych: A (4), B (3), C (2), D (5). Użyj odpowiednich wartości dla każdej zmiennej i narysuj je na wykresie radarowym przy użyciu Matplotlib.

```python
from math import pi

categories = ['A', 'B', 'C', 'D']
values = [4, 3, 2, 5]

N = len(categories)

angles = [n / float(N) * 2 * pi for n in range(N)]
values += values[:1]
angles += angles[:1]

ax = plt.subplot(111, polar=True)
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

ax.plot(angles, values)
ax.fill(angles, values, 'b', alpha=0.1)

plt.title('Wykres radarowy')
plt.show()
```

14. Narysuj wykres z danymi czasowymi przy użyciu dat od stycznia do kwietnia oraz wygenerowanych skumulowanych wartości. Użyj Pandas do stworzenia dat oraz NumPy do generowania skumulowanych danych.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generowanie dat od stycznia do kwietnia
start_date = '2023-01-01'
end_date = '2023-04-30'
dates = pd.date_range(start=start_date, end=end_date, freq='D')

# Generowanie skumulowanych wartości
np.random.seed(42)  # Ustawienie ziarna dla powtarzalności
values = np.cumsum(np.random.rand(len(dates)))

# Tworzenie DataFrame
data = pd.DataFrame({'Date': dates, 'Cumulative Values': values})

# Rysowanie wykresu
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Cumulative Values'], label='Skumulowane wartości', color='blue')
plt.title('Wykres danych czasowych (styczeń-kwiecień)', fontsize=14)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Skumulowane wartości', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

15. Narysuj wykres rozrzutu dla dwóch zmiennych, a następnie dodaj linię regresji, aby zobaczyć zależność między nimi. Użyj losowych danych do wygenerowania wartości dla obu zmiennych. Zastosuj bibliotekę Seaborn do narysowania wykresu oraz linii regresji. Dane: Wygeneruj 100 losowych punktów dla zmiennej X w zakresie od 0 do 10. Wygeneruj zmienną Y jako funkcję X z dodanym szumem (np. Y = 2X + szum Y=2X+szum, gdzie szum to losowe wartości z rozkładu normalnego).

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.uniform(0, 10, 100)
noise = np.random.normal(0, 1, 100)  # Szum
y = 2 * x + noise

data = pd.DataFrame({'X': x, 'Y': y})

sns.regplot(x='X', y='Y', data=data)
plt.title('Wykres rozrzutu z linią regresji')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

16. Narysuj wykres rozrzutu dla zbioru danych Iris, przedstawiający długość działki kielicha w stosunku do długości płatka. Użyj koloru do oznaczenia różnych gatunków.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Załaduj zbiór danych Iris
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target

# Mapowanie nazw gatunków
species_map = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
data['species'] = data['species'].map(species_map)

# Tworzenie wykresu rozrzutu
plt.figure(figsize=(10, 6))
for species, color in zip(['Setosa', 'Versicolor', 'Virginica'], ['red', 'blue', 'green']):
    subset = data[data['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['petal length (cm)'], label=species, color=color)

# Dodanie tytułu i etykiet
plt.title('Wykres rozrzutu: Sepal Length vs Petal Length', fontsize=14)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)
plt.legend(title='Gatunek')
plt.grid(True)
plt.show()
```
