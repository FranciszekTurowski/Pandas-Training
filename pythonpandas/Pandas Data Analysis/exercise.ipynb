```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    """ Task 1: Load the data and display basic information about it."""
    
    data = pd.read_csv('telecom_churn.csv')
    print("TASK 1")
    print(data.info())
    
    """ Task 2: Displays the first 5 rows of the dataset. """
    print("TASK 2") 
    print(data.head())
    
    
    """ Task 3: Display basic statistics."""
    
    print("TASK 3")
    print(data.describe())
   
    """ Task 4: Count the number of churned customers."""
    print("TASK 4")
    print(data['Churn'].value_counts())
   
    """ Task 5: Calculate the average daily minutes for all customers."""
    print("TASK 5")
    print(data['Total day minutes'].mean())
    
    """ Task 6: Calculate the total revenue from daily services."""
    print("TASK 6")
    print(data['Total day charge'].sum())
  
    """ Task 7: Calculate the percentage of customers with an international plan."""
    print("Task 7")
    print(data.columns)
    print((data['International plan'] == 'Yes').mean()*100)
   
    """ Task 8: Calculate the average number of service calls."""
    print("Task 8")
    print(data['Customer service calls'].mean())
    
    """ Task 9: Display the correlation matrix for numerical data."""
    print("TASK 9 ")
    print(data.select_dtypes(include='number'))
   
    """ Task 10: Plot the distribution of churned customers."""
    print("TASK 10")
    data['Churn'].value_counts().plot(kind='bar')
    plt.title('Rozkad klientw - Churn')
    plt.xlabel('Churn')
    plt.ylabel('Liczba klientw')
    plt.show()
  
    """ Task 11: Calculate the average charge per call."""
    print("TASK 11")
    print((data['Total day charge'] / data['Total day calls']).mean())
  
    """ Task 12: Find the customer with the highest daily usage."""
    print("Task 12")
    print(data.loc[data['Total night calls'].idxmax()])
   
    """ Task 13: Filter customers with a voicemail plan."""
    
    print(data[data['Voice mail plan'] == 'Yes'])
    
    """ Task 14: Calculate the average night minutes for churned customers."""
    churned = data[data['Churn'] == True]
    print(churned['Total night minutes'].mean())
   
    """ Task 15: Count the number of unique area codes."""

    #Zliczanie unikalnych kodw obszarw
    print(data['Area code'].nunique())
   
    """Task 16: Plot the distribution of total day minutes."""
    print("TASK 16")
    data['Total day minutes'].value_counts().plot(kind='hist')
    plt.title('Rozkad  total day minutes')
    plt.xlabel('Total day minutes')
    plt.ylabel('Liczba klientw')
    plt.show()
    
   
    """ Task 17: Find customers with more than 5 service calls."""
    print('Task 17')
   
    print(data.loc[data['Customer service calls'] > 5])
    
    """ Task 18: Calculate the churn rate as a percentage."""
    
    
    lost_customers = data['Churn'].sum()
    total_customers = len(data)
    churn_rate = (lost_customers / total_customers) * 100
    print(churn_rate)
    
    """ Task 19: Calculate the average international minutes per customer."""
    print(data['Total intl minutes'].sum() / len(data))
    
    """ Task 20: Find the top 3 states with the highest churn rate."""
   
    
    churn_rate = data.groupby('State')['Churn'].mean()
    top_states = churn_rate.sort_values(ascending=False).head(3)
    print("3 stany o najwyszym wskaniku rezygnacji:\n", top_states)
    
    """ Task 21: Plot the average daily charge by state."""
    print('TASK 21')
    average_charge_by_state = data.groupby('State')['Total day charge'].mean()
    average_charge_by_state.plot(kind='pie')
    plt.title('rednia dzienna opata wedug stanu')
    plt.xlabel('Stan')
    plt.ylabel('rednia dzienna opata')
    plt.show()
    
    """ Task 22: Identify high-value customers based on spending and activity."""
    TiCalls = data['Total intl calls']
    TiCharge = data['Total intl charge']
    average_Total_intl_charge = data['Total intl charge'].mean()
    average_Total_intl_calls  = data['Total intl calls'].mean()
    high_value_customer = (TiCalls > average_Total_intl_calls) &(TiCharge > average_Total_intl_charge)
    print(high_value_customer)
    
    print("avg intl calls: ",average_Total_intl_calls)
    print('------------------------------')
    print(TiCharge)
    print("avg intl charge: ",average_Total_intl_charge)
    vuluable_customers = data[(data['Total intl charge'] > average_Total_intl_charge) | (data['Total intl calls']) > average_Total_intl_calls]
    print("Wartociowi klienci:\n", vuluable_customers)
    print(data.columns)
    
    
    """ Task 23: Analyze service usage patterns."""
    
    data['Total day charge'].value_counts().plot(kind='bar')
    data['Total day minutes'].value_counts().plot(kind='bar')
    plt.title('Rozkad  total day minutes')
    plt.xlabel('Total day minutes')
    plt.ylabel('Liczba klientw')
    plt.show()
   
   
    
    """ Task 24: Identify at-risk customers based on the number of service calls."""
    total_intl_charge = data['Total intl charge']
    filtered_total_intl_charge =  data['Total intl charge'].mean() > total_intl_charge
    account_length = data['Account length']
    mean_account_length =  data['Account length'].mean()
    filtered_data = data[(data['Churn'] == False) 
                         & (mean_account_length < account_length) 
                         & filtered_total_intl_charge]
    print(filtered_data)
  
    print("TASK 24")
    
    """ Task 25: Calculate the average total charge for churned customers."""
    churned_customers = data[data['Churn']== True]
    average_total_charge = churned_customers['Total intl charge'].mean()
    print(f'Average Total Charge for Churned Customers: {average_total_charge:.2f}')
    
    
    """ Task 26: Calculate the average total charge for loyal customers."""
    
    print("TASK 26")
    non_churned_customers = data[data['Churn']== False]
    average_total_charge = non_churned_customers['Total intl charge'].mean()
    print(f'Average Total Charge for Non Churned Customers: {average_total_charge:.2f}')
    
    
    """ Task 27: Calculate the average total charge for customers with an international plan."""
    
    print("TASK 27")
    
    international_plan = data[data['International plan'] == 'Yes']
    avg_total_chage1 = international_plan['Total day charge'].mean()
    print(avg_total_chage1)
    
    """ Task 28: Calculate the average total charge for customers without an international plan."""
    
    print("TASK 28")
    no_international_plan = data[data['International plan'] == 'No']
    avg_total_chage2 = no_international_plan['Total day charge'].mean()
    print(avg_total_chage)
    
    """ Task 29: Calculate the average total charge for customers with voicemail."""
    
    print("TASK 29")
    
    print(data.columns)
    
    customers_with_voicemail = data[data['Voice mail plan'] == 'Yes']
    avg_total_chage3 = customers_with_voicemail['Total day charge'].mean()
    print(avg_total_chage3)
    
    """ Task 30: Find the customer with the highest total charge."""
    
    print("TASK 30")
    
    print(data.loc[data['Total day charge'].idxmax()])
    
    
    average_spend_by_status = data.groupby('Status')['Monthly Spend'].mean()
    print(average_spend_by_status)

    print(data.columns)
    print(data['Customer service calls'].describe())
    print(data['Total intl charge'].describe())
    
  
    
if __name__ == "__main__":
   main()
```
